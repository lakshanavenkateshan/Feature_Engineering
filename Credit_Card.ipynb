```python
#### Independent and Dependent Features
X=df.drop("Class",axis=1)
y=df.Class
```

###Cross Validation Like KFOLD and Hyperparameter Tuning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import KFold
import numpy as np
from sklearn.model_selection import GridSearchCV
```

```python
10.0 **np.arange(-2,3)
```

```python
log_class=LogisticRegression()
grid={'C':10.0 **np.arange(-2,3),'penalty':['l1','l2']}
cv=KFold(n_splits=5,random_state=None,shuffle=False)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Use 'liblinear' solver to support both 'l1' and 'l2'
log_class = LogisticRegression(solver='liblinear', max_iter=1000)

# Grid of hyperparameters
grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.1, 1, 10]
}

# GridSearchCV
clf = GridSearchCV(log_class, grid, cv=cv, n_jobs=-1, scoring='f1_macro')
clf.fit(X_train, y_train)
```

```python
y_pred=clf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

###Under Sampling

```python
from collections import Counter
Counter(y_train)
```

```python
from collections import Counter
from imblearn.under_sampling import NearMiss
ns = NearMiss(sampling_strategy=0.8)
X_train_ns, y_train_ns = ns.fit_resample(X_train, y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_ns)))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train_ns,y_train_ns)
```

```python
y_pred=classifier.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

###Over Sampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
from collections import Counter
from imblearn.over_sampling import RandomOverSampler

# Define oversampler
os = RandomOverSampler(sampling_strategy=0.75)

# Apply oversampling
X_train_ns, y_train_ns = os.fit_resample(X_train, y_train)

# Print class distribution
print("The number of classes before fit: {}".format(Counter(y_train)))
print("The number of classes after fit: {}".format(Counter(y_train_ns)))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train_ns,y_train_ns)
```

```python
y_pred=classifier.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
