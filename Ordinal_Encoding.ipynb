COUNT OR FREQUENCY ENCODING

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()
```

```python
columns=[1,3,5,6,7,8,9,13]
```

```python
train_set=train_set[columns]
```

```python
train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns[:]:
    print(feature,":",len(train_set[feature].unique()),'labels')
```

This loop is used to check how many unique values (or distinct labels) each column in the dataset contains. Here, train_set.columns[:] iterates through all the column names in the DataFrame. For each column (stored in the variable feature), train_set[feature].unique() finds all the unique values present in that column, and len(...) counts how many of those unique values exist. Finally, the print statement displays the column name, followed by the number of unique labels it has. In simple terms, this code helps you quickly understand the variety of values in each column, which is especially useful in tasks like data preprocessing, feature engineering, or when analyzing categorical features in a dataset.

```python
country_map=train_set['Country'].value_counts().to_dict()
#country_map is a dictionary that maps each country to the number of times it occurs in your dataset.
```

```python
train_set['Country']=train_set['Country'].map(country_map)
train_set.head(20)
```

###Advantages
Easy To Use.Not increasing feature space.
###Disadvantages
It will provide same weight if the frequencies are same

```python
import pandas as pd
df=pd.read_csv('train (1).csv', usecols=['Cabin','Survived'])
df.head()
```

```python
df['Cabin'] = df['Cabin'].fillna('Missing')
#All missing values in the Cabin column are replaced with "Missing"
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
#The Cabin column will now only contain the first letter of the cabin code (which usually indicates the deck on the ship).
#For missing values (NaN), since they were turned into the string "nan", the first letter will be "n".
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
enumerate(ordinal_labels,0)
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
#enumerate() gives you both an index and the value while looping through a list.
#{k: i for i, k in ... }
#This is a dictionary comprehension.
#It creates keyvalue pairs where:
#k = the label (like 'Poor')
#i = the index number (like 0)
#ordinal_labels2 is a dictionary that assigns a numeric code to each label in ordinal_labels.
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)
df.head()
```

#MEAN ENCODING

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)
df.head()
```
